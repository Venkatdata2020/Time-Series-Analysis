```
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import os
from datetime import datetime
import pandas as pd

mpl.rcParams["figure.figsize"] =(10,8)
mpl.rcParams["axes.grid"] = False
```

```
def parse(x):
  return datetime.strptime(x,"%m/%d/%Y")
```

```
df = pd.read_csv("electricity_consumption.csv",parse_dates=['Bill_Date'],date_parser=parse)
df.head()
```

```
bill_df = df.set_index('Bill_Date')
```

```
bill_df.head()
```

```
bill_2018 = bill_df["2016":"2018"][["Billed_amount"]]
bill_2018.head()
```

```
bill_2018['Billed_amount'].rolling(window=3).mean()
```

```
bill_2018['ma_rolling_3'] = bill_2018['Billed_amount'].rolling(window=3).mean().shift(1)
```

```
bill_2018
```

```
bill_2018.plot()
```

```
def wma(weights):
  def calc(x):
    return (weights*x).mean()
  return calc
```

```
bill_2018['Billed_amount'].rolling(window=3).apply(wma(np.array([0.5,1,1.5])))
```

```
bill_2018['wma_rolling_3']=bill_2018['Billed_amount'].rolling(window=3).apply(wma(np.array([0.5,1,1.5]))).shift(1)
```

```
bill_2018.head()
```

```
bill_2018.plot()
```

```
bill_2018["ewm_window_3"] = bill_2018["Billed_amount"].ewm(span=3,adjust=False,min_periods=0).mean().shift(1)
```

```
bill_2018
```

```
bill_2018.plot()
```

```
bill_2018['ewm_window_3_7']=bill_2018['Billed_amount'].ewm(alpha=0.7,adjust=False,min_periods=3).mean().shift(1)
```

```
bill_2018.plot()
```

```
bill_2018['ewm_window_3_3']=bill_2018['Billed_amount'].ewm(alpha=0.3,adjust=False,min_periods=3).mean().shift(1)
```
